# **Matplotlib**

For the Python programming language, Matplotlib is a popular 2D charting toolkit. It gives you access to a wide range of tools for using Python to build various kinds of static, animated, and interactive visualizations. Matplotlib is a well-liked option for data visualization, scientific charting, and producing publication-quality visuals because of its user-friendly, adaptable, and highly adjustable design.

## **01. Line Graph**
* Line chart/graph is used whenever we want to show the Trend/Rate of a particular element be it stocks, sugar levels etc.
* Whenever time component is there it should always be in the X-axis and Line Chart should be prefered

```python
x=np.random.randint(low=-5,high=26,size=(20,))
y=np.random.randint(low=-7,high=36,size=(20,))
x.sort(), y.sort()

#x=[1,2,3];y=[10,20,30]

plt.title("My Line Graph ",loc="center")  #Set a title for the Axes.
plt.xlabel("Time")
plt.ylabel("Development")

plt.plot(x,y)
```

```python
x=np.random.randint(low=-5,high=26,size=(20,))
y=np.random.randint(low=-7,high=36,size=(20,))
x.sort(), y.sort()

plt.title("My Line Graph ",loc="center")  #Set a title for the Axes.
plt.xlabel("Time")
plt.ylabel("Development")

plt.plot(x,y,color='c',linewidth=1.5,linestyle='-.',marker="*",markersize=5.5)   #X-axis followed by Y-Axis
plt.grid()
plt.show()  #It hides the metadata of the graph
```

### **Creating sub-plots in a given ploting area region**

```python
x=np.linspace(1,10)

plt.figure(figsize=(10,10),edgecolor="b")

plt.subplot(2,2,1)  #Creating 4 regions in One Plot (2 x2) and Activating the 1st Region to use for
plt.plot(x,np.log(x),color='c',linewidth=1.7,linestyle="--",markersize=5.5,label="Log Line")
plt.legend()  #This adds the Label of the respective lines in the Graph Plot
plt.grid() #Will show a grid representation of the plot

plt.subplot(2,2,2)  #Activating the 2nd Region
plt.plot(x,np.sqrt(x),color='y',linewidth=1.7,linestyle=":",markersize=5.5,label="Sqrt Line")
plt.legend()  #This adds the Label of the respective lines in the Graph Plot
plt.grid()  #Will show a grid representation of the plot

plt.subplot(2,2,3)  #Activating the 3rd Region
plt.plot(x,np.tan(x),color='r',linewidth=1.7,linestyle="-",markersize=5.5,label="Tan Line")
plt.legend()  #This adds the Label of the respective lines in the Graph Plot
plt.grid()  #Will show a grid representation of the plot

plt.subplot(2,2,4)  #Activating the 4th Region
plt.plot(x,np.cos(x),color='m',linewidth=1.7,linestyle="-",markersize=5.5,label="Cos Line")

plt.legend()  #This adds the Label of the respective lines in the Graph Plot
plt.grid()  #Will show a grid representation of the plot
plt.show()  # On using plt.show() it hides the metadata of the graph
```

## **02.Histogram**

* Mainly used to show the distribution of a continous random varaible across a given range of upper and lower limit
* **X-Axis**->Represents the bins, which are the different ranges of values your data falls into (e.g., apple size categories).[3-4][5-6][7-8]..etc
* **Y-Axis**->Represents the frequency or count. It shows how many data points fall within each bin (e.g., number of apples in each size category).
**The plt.hist() function in Matplotlib creates a histogram, which is a visual way to show how your data (in this case, the Fare column) is distributed.**

```python
df=pd.read_csv("Titanic_Dataset.csv")
print(df.info())
```

```python
df["Fare"].aggregate(["max","min","count"])
```

```python
(df.sample(5))
```

```python
df["Fare"].value_counts()
```

```python
plt.hist(df["Fare"],bins=np.linspace(0,513,50),color='c',density=True)
plt.xticks(np.arange(0, 512,25))  # Ticks every 25 units
# density = True It changes the y-axis from showing raw counts to showing probability density.
plt.show()
```

### **2.1 Stacked Histogram**

```python
data1,data2=np.random.standard_normal(size=50),np.random.standard_normal(size=50)
plt.hist([data1,data2],bins=20,density=False,color=['r','b'],label=["data1","data2"])
plt.legend()
plt.show()
```

## **04. Bar Graph**

* It is used to show the **distribution by-count** of a categorical data

```python
plt.bar(x=("A","B","C"),height=(45,12,68))
```

```python
bank.sort_values(by="Age",inplace=True)
bank.sample(5)
```

```python
se=bank["Age"].value_counts()
plt.figure(figsize=(10,10))
plt.xticks(np.arange(10,100,5))
plt.yticks(np.arange(10,490,))
plt.bar(data=se,x=se.index,height=se.values,color='g')
```

## **05. Pie Chart**

* This ploting visual show the distribution of a certain Category/Label in its %age

```python
plt.figure(figsize=(15,15))
plt.pie(data=se,x=se.values,labels=se.index,autopct="%1.1f%%",shadow=True)
plt.show()
```
