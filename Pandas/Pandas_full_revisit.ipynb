#### **Creating Start and End Time Stamps**

```python
def Fibo(n):
    result=[0,1]
    if(n<=2 and n!=0):
        return result[:n]
    elif(n>3):
        for i in range(3,n+1):
            result.append(result[i-3]+result[i-2])
        return result
```
